```python
import sys
import os
import time

import h5py
import matplotlib.pyplot as plt
import numpy as np
import palettable

from Setup import generatePermIndices, SecondOrderCPA

ATTACK_N_A = 400000

def align_traces(traces, reference_trace=None):
    
    if reference_trace is None:
        reference_trace = np.mean(traces, axis=0)
    
    aligned_traces = np.array([np.roll(trace, np.argmax(np.correlate(trace, reference_trace, mode='full')) - (len(trace) - 1)) for trace in traces])
    
    return aligned_traces


all_c_range = np.concatenate([np.arange(5550 + 373 * i, 5561 + 373 * i) for i in range(16)])

DATA_FILE = "ascadv2-extracted.h5"
file = h5py.File(DATA_FILE, "r")

c_pick = [5552, 5553]
traces = file["Profiling_traces"]["traces"][0:ATTACK_N_A, c_pick]  

aligned_traces = align_traces(traces)

rout_pick = [338, 339, 341, 342, 343, 345]
traces_rout = file["Profiling_traces"]["traces"][0:ATTACK_N_A, rout_pick]  

aligned_traces_rout = align_traces(traces_rout)

meta = file["Profiling_traces"]["metadata"][0:ATTACK_N_A]
labels = file["Profiling_traces"]["labels"][0:ATTACK_N_A]
masks = meta["masks"][0:ATTACK_N_A]

inds = np.array([generatePermIndices(np.arange(16), *masks[i][:4]) for i in range(ATTACK_N_A)])[:, 0]

plains = np.array([meta["plaintext"][i, inds[i]] for i in range(ATTACK_N_A)])
keys = np.array([meta["key"][i, inds[i]] for i in range(ATTACK_N_A)])

alphas = labels["alpha_mask"].flatten()
betas = labels["beta_mask"].flatten()

print(f"shape of plains: {plains.shape} shape of keys: {keys.shape} shape of alphas: {alphas.shape} shape of betas: {betas.shape}")

start_time = time.time()

my_CPA_2_ = SecondOrderCPA(traces=aligned_traces, traces_rout=aligned_traces_rout, plains=plains, keys=keys, alphas=alphas)
ge, succ_rate, corr_k_t = my_CPA_2_.compute_guessing_entropy()

end_time = time.time()
print("CPA running time: %s seconds" % (end_time - start_time))

print('guessing entropy:', ge)
print('success rate:', succ_rate)
print('max correlation:', corr_k_t[-10:])

plt.rcParams['figure.figsize'] = [16, 4]
plt.rc('font', family='Arial', weight='bold', size='20')

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 14), dpi=500)

ax1.set_yscale('log', base=2)
ax1.set_ylim(0.4, 600)
curr_Na = np.arange(ge.shape[0])
ax1.set_xlabel("Number of Power Traces Analysed")
ax1.set_ylabel("Guess Entropy")
line0 = ax1.plot(curr_Na, ge, label="Guess Entropy", color="red")

ax1_right = ax1.twinx()
ax1_right.set_ylabel("Success Rate")
ax1_right.set_ylim(-0.1, 1.1)
line1 = ax1_right.plot(curr_Na, succ_rate, label="Success Rate", color="blue")

ax1.legend([line0[0], line1[0]], ["Guess Entropy", "Success Rate"], fancybox=True, loc="center right")
ax1.grid(True)

ax2.set_ylim(0, 1.0)
ax2.set_xlabel("Number of Power Traces Analysed")
ax2.set_ylabel("Correlation Coefficient")
ax2.plot(corr_k_t, label="Correlation Coefficient $\\rho_{ck,tk}$", color="blue")
ax2.legend(fancybox=True)
ax2.grid(True)

plt.tight_layout()
plt.savefig("CPA2_analysis.jpg")
plt.show()
```
